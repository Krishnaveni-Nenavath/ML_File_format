```python
frames = cv2.VideoCapture(r"C:\Users\krish\Documents\identify_hut_Palace\hut.mp4")  # converting any videos to frames 
while True: 
    suc,img = frames.read()  # if the frames present than return the booleam true or else False
    if suc == False: 
        break
    
    cv2.imshow("vid",img) 
    if cv2.waitKey(1) & 255 == ord("C"): # comes out ffrom the window at any  time after we give the input  
        break
cv2.destroyAllWindows()
    
```

```python
frames = cv2.VideoCapture(r"C:\Users\krish\Documents\identify_hut_Palace\hut.mp4")  # converting any videos to frames 
while True: 
    suc,img = frames.read()  # if the frames present than return the booleam true or else False
    if suc == False: 
        break
    b,g,r = cv2.split(img) 
    z = np.zeros(b.shape,dtype = np.uint8) 
    cv2.imshow("blue",cv2.merge([b,z,z])) 
    cv2.imshow("green",cv2.merge([z,g,z])) 
    cv2.imshow("RED",cv2.merge([z,z,r]))
    cv2.imshow("vid",img) 
    if cv2.waitKey(1) & 255 == ord("C"): # comes out ffrom the window at any  time after we give the input  
        break 
cv2.destroyAllWindows()
    
```
